# Diccionario **Ilustrador**

```python
Diccionario_ilustrador = {}

def agregar_ilustrador():

    id_ilustrador = input("ID: ")
    nombre = input("Nombre: ")
    especialidad = input("Especialidad: ")
    pais = input("País: ")
    ilustrador = {
        "ID": id_ilustrador,
        "Nombre": nombre,
        "Especialidad": especialidad,
        "Pais": pais
    }
    Diccionario_ilustrador.append(ilustrador)
    print("\nIlustrador agregado correctamente\n")
    listar_ilustrador()
def listar_ilustrador():
    if len(Diccionario_ilustrador) == 0:
        print("\nNo hay ilustradores registrados\n")
    else:
        print("\n================ LISTA DE ILUSTRADORES ================\n")
        print("{:<10} {:<20} {:<20} {:<15}".format(
            "ID", "NOMBRE", "ESPECIALIDAD", "PAIS"
        ))
        print("-" * 70)
        for ilustrador in Diccionario_ilustrador:
            print("{:<10} {:<20} {:<20} {:<15}".format(
                ilustrador["ID"],
                ilustrador["Nombre"],
                ilustrador["Especialidad"],
                ilustrador["Pais"]
            ))
        print()
def buscar_ilustrador():
    buscar = input("Ingrese el nombre del ilustrador a buscar: ")

    encontrado = False

    for ilustrador in Diccionario_ilustrador:

        if ilustrador["Nombre"] == buscar:

            print("\n=========== ILUSTRADOR ENCONTRADO ===========\n")
            for key, value in ilustrador.items():
                print(key, ":", value)
            encontrado = True
    if encontrado == False:
        print("\nIlustrador no encontrado\n")
    listar_ilustrador()

def borrar_ilustrador():
    borrar = input("Ingrese el nombre del ilustrador a borrar: ")
    for ilustrador in Diccionario_ilustrador:
        if ilustrador["Nombre"] == borrar:
            Diccionario_ilustrador.remove(ilustrador)
            print("\nIlustrador eliminado correctamente\n")
            listar_ilustrador()
            return
    print("\nIlustrador no encontrado\n")

def actualizar_ilustrador():
    actualizar = input("Ingrese el nombre del ilustrador a actualizar: ")
    for ilustrador in Diccionario_ilustrador:
        if ilustrador["Nombre"] == actualizar:
            ilustrador["Nombre"] = input("Nuevo nombre: ")
            ilustrador["Especialidad"] = input("Nueva especialidad: ")
            ilustrador["Pais"] = input("Nuevo país: ")
            print("\nIlustrador actualizado correctamente\n")
            listar_ilustrador()
            return
    print("\nIlustrador no encontrado\n")

while True:
    print("\n============= MENÚ ============")
    print("1. Agregar ilustrador en un diccionario")
    print("2. Listar diccionarios de ilustradores")
    print("3. Buscar ilustrador en diccionario")
    print("4. Borrar ilustrador en diccionario")
    print("5. Actualizar ilustrador en diccionario")
    print("6. Salir")
    opcion = input("Seleccione una opción: ")
    if opcion == "1":
        agregar_ilustrador()

    elif opcion == "2":
        listar_ilustrador()

    elif opcion == "3":
        buscar_ilustrador()

    elif opcion == "4":
        borrar_ilustrador()

    elif opcion == "5":
        actualizar_ilustrador()

    elif opcion == "6":
        print("\nPrograma finalizado")
        break

    else:
        print("\nOpción incorrecta\n")
print("Programa realizado por Oscar Flores NL = 19")
```
